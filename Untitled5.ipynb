Step 1. Import the necessary libraries

Step 4. Discover what is the mean age per occupation
Step 5. Discover the Male ratio per occupation and sort it from the most to the least Step 6. For each occupation, calculate the minimum and maximum ages
Step 7. For each combination of occupation and sex, calculate the mean age
Step 8. For each occupation present the percentage of women and men

```python
# importing phython libraries

#Importing the pandas lib as pd

import pandas as pd
```

```python
# Step 2: Bringing the dataset in from the specified location

given_url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
```

```python
# Step 3 Assign to a variable called users
users = pd.read_csv(given_url, sep='|')# The data is separated by '|' to match the file's structure
users.head()
```

```python

# Step 4. Discover what is the mean age per occupation

# Finding unique occupations
occupations = users['occupation'].unique()

print("Mean Age per Occupation:")
print("-" * 40)
print(f"{'Occupation':<20} | {'Mean Age':<8}")
print("-" * 40)

# Iterating through each occupation and filtering rows for calculation
for occupation in occupations:
    age_mean = users[users['occupation'] == occupation]['age'].mean()
    print(f"{occupation:<20} | {age_mean:.2f} years")
```

```python
# Step 5. Discover the Male ratio per occupation and sort it from the most to the least

# Using crosstab to calculate the male ratio per occupation
gender_counts = pd.crosstab(users['occupation'], users['gender'], normalize='index')

# Selecting the male ratio column and sorting
male_ratio_per_occupation_sorted = gender_counts['M'].sort_values(ascending=False)

# Printing the results
print("Male Ratio per Occupation (Sorted):")
print("-" * 50)
print(f"{'Occupation':<20} | {'Male Ratio (%)':<15}")
print("-" * 50)
for occupation, ratio in male_ratio_per_occupation_sorted.items():
    print(f"{occupation:<20} | {ratio * 100:.2f}%")
```

```python
# Step 6. For each occupation, calculate the minimum and maximum ages

# Using describe to get statistics per occupation
stats_per_occupation = users.groupby('occupation')['age'].describe()

# Extracting min and max ages manually
min_max_age_per_occupation = stats_per_occupation[['min', 'max']]

# Printing the results
print("Minimum and Maximum Ages per Occupation:")
print("-" * 70)
print(f"{'Occupation':<20} | {'Min Age':<8} | {'Max Age':<8}")
print("-" * 70)
for occupation, ages in min_max_age_per_occupation.iterrows():
    print(f"{occupation:<20} | {ages['min']:<8} | {ages['max']:<8}")
```

```python
# Step 7. For each combination of occupation and sex, calculate the mean age

# Define a custom function to calculate mean age
def calculate_mean_age(group):
    return group['age'].mean()

# Applying the custom function to each group of occupation and gender
mean_age_per_occupation_sex = users.groupby(['occupation', 'gender']).apply(calculate_mean_age).reset_index(name='Mean Age')

# Printing the results with a clean format
print("Mean Age per Occupation and Gender:")
print("-" * 50)
print(f"{'Occupation':<20} | {'Gender':<5} | {'Mean Age':<8}")
print("-" * 50)

# Using a simple for loop to print the results
for _, row in mean_age_per_occupation_sex.iterrows():
    print(f"{row['occupation']:<20} | {row['gender']:<5} | {row['Mean Age']:<8.2f}")
```

```python
# Step 8. For each occupation, present the percentage of women and men

# Step 1: Calculate the gender distribution as a percentage within each occupation
def gender_percentage(group):
    total_count = group.size
    return group.value_counts() / total_count * 100

# Step 2: Apply the custom function to calculate percentages for each group
gender_percentage_per_occupation = users.groupby('occupation')['gender'].apply(gender_percentage).reset_index(name='Percentage (%)')

# Step 3: Rename columns for clarity
gender_percentage_per_occupation.columns = ['Occupation', 'Gender', 'Percentage (%)']

# Step 4: Round the percentage values to two decimal places
gender_percentage_per_occupation['Percentage (%)'] = gender_percentage_per_occupation['Percentage (%)'].round(2)

# Printing the final result
print(gender_percentage_per_occupation)
```

```python
import pandas as pd

# Step 2: Read the file
data = pd.read_fwf('wind.txt')

# Step 3: Rename columns
data.columns = ['Yr', 'Mo', 'Dy', 'RPT', 'VAL', 'ROS', 'KIL', 'SHA', 'BIR', 'DUB', 'CLA', 'MUL', 'CLO', 'BEL', 'MAL']

# Step 3: Create the Date column and format it
data['Date'] = pd.to_datetime(data[['Yr', 'Mo', 'Dy']].astype(str).agg('-'.join, axis=1), format='%y-%m-%d')

# Set the 'Date' column as the index and drop the original columns
data.set_index('Date', inplace=True)
data.drop(['Yr', 'Mo', 'Dy'], axis=1, inplace=True)

# Step 4: Define a function to adjust years
def correct_years(df):
    # Adjust years based on a conditional check
    df.index = df.index.map(lambda x: x - pd.DateOffset(years=100) if x.year >= 2060 else x)
    return df

# Apply the function to the DataFrame
data = correct_years(data)

# Display the resulting DataFrame
print(data)
```

```python
# Step 6: Identify and summarize missing values

# Get a DataFrame showing missing values per column
missing_values_df = data.isna().sum().reset_index()
missing_values_df.columns = ['Location', 'Missing Values']

# Calculate the total number of missing values
total_missing_values = missing_values_df['Missing Values'].sum()

# Print missing values per location
print("Missing values per location:")
print(missing_values_df)

# Print the total number of missing values
print("\nTotal missing values:", total_missing_values)
```

```python
# Step 7

print(f"Total non-missing values: {np.prod(data.shape) - data.isnull().sum().sum()}")
     
```

```python
# Step 8: Calculate and display the mean windspeed

# Calculate the mean windspeed for each column (location)
mean_windspeed_per_location = data.mean()

# Calculate the overall mean windspeed across all locations and times
overall_mean_windspeed = mean_windspeed_per_location.mean()

# Print the overall mean windspeed
print("Mean windspeed over all locations and times:", overall_mean_windspeed)

# Print the rounded mean windspeed
print("After rounding off:", round(overall_mean_windspeed))
```

```python
# Step 9: Calculate and display statistics for each location

# Calculate statistics for each location
loc_stats = data.describe().loc[['min', 'max', 'mean', 'std']]

# Print the statistics
print("Location statistics:")
print(loc_stats)
```

```python
# Step 10: Calculate daily statistics

# Calculate statistics for each day (row-wise)
day_stats = pd.DataFrame({
    'min': data.min(axis=1),
    'max': data.max(axis=1),
    'mean': data.mean(axis=1),
    'std': data.std(axis=1)
})

# Print the resulting DataFrame
print("Daily statistics:")
print(day_stats)
```

```python
# Step 11: Calculate the average windspeed in January for each year

# Filter data for January
data_january = data[data.index.month == 1]

# Create a pivot table to calculate the average windspeed for each year and location
average_january_windspeed = pd.pivot_table(data_january, 
                                          index=data_january.index.year, 
                                          aggfunc='mean')

# Print the results
print("Average windspeed in January for each location:")
print(average_january_windspeed)
```

```python
# Step 12: Resample data to yearly frequency and calculate the mean

# Resample the data to yearly frequency, aggregating with mean
yearly_data = data.resample('Y').mean()

# Print the results
print("Yearly downsampled data:")
print(yearly_data)
```

```python
# Step 13: Resample data to monthly frequency and calculate the mean

# Resample the data to monthly frequency, aggregating with mean
monthly_data = data.resample('M').mean()

# Print the results
print("Monthly downsampled data:")
print(monthly_data)
```

```python
# Step 14: Resample data to weekly frequency and calculate the mean

# Resample the data to weekly frequency, aggregating with mean
weekly_data = data.resample('W').mean()

# Print the results
print("Weekly downsampled data:")
print(weekly_data)
```

Create a scatter plot showing the relationship between the total revenue earned by arcades and the number of Computer Science PhDs awarded in the U.S. between 2000 and 2009. Don't forget to label your axes! Color each dot according to its year.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the data
data = pd.read_csv('arcade-revenue-vs-cs-doctorates.csv')

# Set up the figure and axis
plt.figure(figsize=[10, 6])

# Create a scatter plot with different colors and markers
scatter = plt.scatter(
    data['Total Arcade Revenue (billions)'],
    data['Computer Science Doctorates Awarded (US)'],
    c=data['Year'],
    cmap='plasma',         # Changed colormap for variation
    s=80,                  # Adjusted size of markers
    alpha=0.9,             # Increased opacity for better visibility
    edgecolor='k'          # Added edge color for markers
)

# Add colorbar with a label
plt.colorbar(scatter, label='Year', orientation='vertical')

# Label the axes and add a title
plt.xlabel('Total Arcade Revenue (billions)', fontsize=12)
plt.ylabel('Computer Science Doctorates Awarded (US)', fontsize=12)
plt.title('Arcade Revenue vs. CS Doctorates Awarded (2000-2009)', fontsize=14, fontweight='bold')

# Add grid lines
plt.grid(True, linestyle='--', alpha=0.6)

# Show the plot
plt.tight_layout()
plt.show()
```
